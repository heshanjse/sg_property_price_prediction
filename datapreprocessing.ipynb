```python
#format month
df4 = pd.read_csv("Historical_CPI.csv")
df4.head()
```

```python
# change the 'Month' column
def modify_month(month_str):
    if month_str.startswith('jan'):
        return f"{month_str[3:]}-01"
    elif month_str.startswith('feb'):
        return f"{month_str[3:]}-02"
    elif month_str.startswith('mar'):
        return f"{month_str[3:]}-03"
    elif month_str.startswith('apr'):
        return f"{month_str[3:]}-04"
    elif month_str.startswith('may'):
        return f"{month_str[3:]}-05"
    elif month_str.startswith('jun'):
        return f"{month_str[3:]}-06"
    elif month_str.startswith('jul'):
        return f"{month_str[3:]}-07"
    elif month_str.startswith('aug'):
        return f"{month_str[3:]}-08"
    elif month_str.startswith('sep'):
        return f"{month_str[3:]}-09"
    elif month_str.startswith('oct'):
        return f"{month_str[3:]}-10"
    elif month_str.startswith('nov'):
        return f"{month_str[3:]}-11"
    elif month_str.startswith('dec'):
        return f"{month_str[3:]}-12"
    else:
        return month_str

# Apply the custom function to the 'Month' column
df4['Month'] = df4['Month'].apply(modify_month)
```

```python
df4.to_csv("historyCPIFormated.csv", encoding='utf-8', index=False)
```
